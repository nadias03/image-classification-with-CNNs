```python
import os
import shutil
from pathlib import Path

SOURCE_DIR = Path("../data")
SUBSET_DIR = Path("../data_subset")

IMAGES_PER_CLASS = 100

for split in ["train", "valid", "test"]:
    for class_name in os.listdir(SOURCE_DIR / split):
        src_class = SOURCE_DIR / split / class_name
        dst_class = SUBSET_DIR / split / class_name
        dst_class.mkdir(parents=True, exist_ok=True)
        
        images = list(src_class.glob("*.png"))[:IMAGES_PER_CLASS]
        for img in images:
            shutil.copy(img, dst_class / img.name)

    print(f"✅ {split} split done")

DATA_DIR = SUBSET_DIR
TRAIN_DIR = DATA_DIR / "train"
VALID_DIR = DATA_DIR / "valid"
TEST_DIR = DATA_DIR / "test"
```
